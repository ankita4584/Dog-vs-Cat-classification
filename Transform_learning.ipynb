Cat vs Dog

Use a pretrained model (like VGG16 or ResNet50) and fine-tune it to classify whether an image contains a cat or dog.

Transform learning Model


```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,Dropout,Lambda
from sklearn.model_selection import train_test_split

train_ds=keras.utils.image_dataset_from_directory(directory='/content/drive/MyDrive/Colab Notebooks/kaggle-dataset/training_set',labels='inferred',label_mode='int',batch_size=32,image_size=(256,  256))

test_ds=keras.utils.image_dataset_from_directory(directory='/content/drive/MyDrive/Colab Notebooks/kaggle-dataset/test_set',labels='inferred',label_mode='int',batch_size=32,image_size=(256,  256))

inputs=keras.Input(shape=(256,256,3))
s = Lambda(lambda x: x / 255)(inputs)
```

Transform learning model

```python
from keras.applications import MobileNetV2, EfficientNetB0

from keras.preprocessing import image
from keras import models,layers
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,Dropout,Lambda

import numpy as np

inputs=keras.Input(shape=(256,256,3))
s = Lambda(lambda x: x / 255)(inputs)
model = MobileNetV2(weights='imagenet',include_top=False,input_tensor=s,input_shape=(256,256,3)) #here imagenet because we take pretrain weight or it if we write a zero it will not be worth it
#here attribute include top is like includeng classification layer is useless

for layer in model.layers:
  layer.trainable=False #here we freeze the weight we are not going to update gradient of it

x=layers.Flatten()(model.output)
x = layers.Dense(128, activation="relu")(x)
output = layers.Dense(1, activation="sigmoid")(x)  # binary: cat vs dog

model = models.Model(inputs=model.input, outputs=output)

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.fit(train_ds,epochs=10,validation_data=test_ds)

print(model.summary())



```
